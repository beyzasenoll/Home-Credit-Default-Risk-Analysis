```python
#this method for finding categorical, numerical and categorical rows with higher cardinality
def grab_col_names(dataframe, cat_th=10, car_th=20):
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    num_but_cat = [col for col in dataframe.columns if
                   dataframe[col].nunique() < cat_th and dataframe[col].dtypes != "O"]
    cat_but_car = [col for col in dataframe.columns if
                   dataframe[col].nunique() < car_th and dataframe[col].dtypes == "O"]

    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cat]

    return cat_cols, num_cols, cat_but_car
```

### Correlation

```python
num_cols=app_train.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr_matrix = num_cols.corr()

# Find correlations with the target and sort
correlations = corr_matrix['TARGET'].sort_values()

# Display correlations
print('Most Positive Correlations:\n', correlations.tail(15))
print('\nMost Negative Correlations:\n', correlations.head(15))
```

```python
# Convert DAYS_BIRTH to positive values
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])

# Plot age distribution
plt.figure(figsize=(3, 3))
plt.hist(app_train['DAYS_BIRTH'] / 365, bins=25, edgecolor='k')
plt.title('Age Distribution of Clients')
plt.xlabel('Age (years)')
plt.ylabel('Count')
plt.show()

```

```python
# Plot KDE of age by target class
plt.figure(figsize=(5, 5))
sns.kdeplot(data=app_train[app_train['TARGET'] == 0], x='DAYS_BIRTH', label='Repaid', fill=False)
sns.kdeplot(data=app_train[app_train['TARGET'] == 1], x='DAYS_BIRTH', label='Not Repaid', fill=False)
plt.xlabel('Age (days)')
plt.ylabel('Density')
plt.title('Density of Age by Loan Repayment Status')
plt.legend()
plt.show()
```

```python
# Create age brackets
age_data = app_train[['TARGET', 'DAYS_BIRTH']]
age_data['YEARS_BIRTH'] = age_data['DAYS_BIRTH'] / 365
age_data['AGE_BIN'] = pd.cut(age_data['YEARS_BIRTH'], bins=np.arange(20, 75, 5))

# Compute average repayment failure by age bracket
age_summary = age_data.groupby('AGE_BIN').mean()

# Plot average failure to repay by age group
plt.figure(figsize=(3, 3))
plt.bar(age_summary.index.astype(str), 100 * age_summary['TARGET'])
plt.xticks(rotation=75)
plt.xlabel('Age Group (years)')
plt.ylabel('Failure to Repay (%)')
plt.title('Failure to Repay by Age Group')
plt.show()
```

The analysis shows that younger clients tend to default on loans more often, with a clear trend indicating higher failure rates among the younger age groups.

Pairs Plot

```python
# Prepare data for pairs plot
plot_data = ext_data.drop(columns='DAYS_BIRTH').copy()
plot_data['YEARS_BIRTH'] = app_train['DAYS_BIRTH'] / 365

# Drop missing values and limit to the first 100,000 rows
plot_data = plot_data.dropna().iloc[:100000]

# Define a function to annotate correlation coefficients
def corr_func(x, y, **kwargs):
    r = np.corrcoef(x, y)[0, 1]
    ax = plt.gca()
    ax.annotate(f"r = {r:.2f}", xy=(0.2, 0.8), xycoords=ax.transAxes, size=14, color='black')

# Create PairGrid object for plotting.
grid = sns.PairGrid(plot_data, hue='TARGET', palette={0: 'blue', 1: 'red'}, height=3)
grid.map_upper(plt.scatter, alpha=0.2)
grid.map_diag(sns.histplot, kde=True)
grid.map_lower(sns.kdeplot, cmap='OrRd', alpha=0.6)
grid.map_lower(corr_func)

plt.suptitle('Pairs Plot of EXT_SOURCE Features and Age', size=20)
plt.show()
```

There does appear to be a moderate positive linear relationship between the EXT_SOURCE_1 and the DAYS_BIRTH.
